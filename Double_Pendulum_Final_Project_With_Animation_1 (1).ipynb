### Setting the Scene

Consider a double pendulum system. The pendulum on top, which we will call Pendulum 1, has length $\mathit{l}_1$ with a mass $\mathit{m}_1$ attached to it. The pendulum attached to it (Pendulum 2) has length $\mathit{l}_2$ with a mass $\mathit{m}_2$ attached to it. The generalized coordinates we are using will be $\phi_1$ and $\phi_2$, which measure the angular displacement of Pendulums 1 and 2, respectively. 

### Deriving the Equations of Motion

To derive the equations of motion, we look to set up the Lagrangian. Let x be the horizontal axis increasing to the right and y be the vertical axis increasing in the upward direction. The total kinetic and potential energies are the sum of the corresponding kinetic and potential energies of Pendulums 1 and 2. 

#### Pendulum 1:
For Pendulum 1, mass 1 has (x, y) coordinates of $(\mathbf{l}_1\sin\boldsymbol{\phi}_1,\,-\mathbf{l}_1\cos\boldsymbol{\phi}_1)$. Since kinetic energy involves velocity, we can differentiate the given (x, y) coordinates to find it.

$\mathbf{v}_1=(\mathbf{l}_1\,\dot{\boldsymbol{\phi}}_1\cos\boldsymbol{\phi}_1,\,-\mathbf{l}_1\,\dot{\boldsymbol{\phi}}_1\sin\boldsymbol{\phi}_1)$

Our kinetic energy for mass 1:

$\mathbf{T}_1=\frac{1}{2}\,\mathbf{m}_1\bigl(\mathbf{l}_1\,\dot{\boldsymbol{\phi}}_1\bigr)^2$


#### Pendulum 2:
Similarly, for Pendulum 2, the (x, y) coordinates are $(\mathbf{l}_1\sin\boldsymbol{\phi}_1+\mathbf{l}_2\sin\boldsymbol{\phi}_2,\,-\mathbf{l}_1\cos\boldsymbol{\phi}_1-\mathbf{l}_2\cos\boldsymbol{\phi}_2)$. Then, the velocity is

$\mathbf{v}_2=(\mathbf{l}_1\,\dot{\boldsymbol{\phi}}_1\cos\boldsymbol{\phi}_1+\mathbf{l}_2\,\dot{\boldsymbol{\phi}}_2\cos\boldsymbol{\phi}_2,\;\mathbf{l}_1\,\dot{\boldsymbol{\phi}}_1\sin\boldsymbol{\phi}_1+\mathbf{l}_2\,\dot{\boldsymbol{\phi}}_2\sin\boldsymbol{\phi}_2)$


Using trigonometric identities, we express the kinetic energy for mass 2.

$\mathbf{T}_2=\frac{1}{2}\,\mathbf{m}_2\Bigl[\bigl(\mathbf{l}_1\,\dot{\boldsymbol{\phi}}_1\bigr)^2+\bigl(\mathbf{l}_2\,\dot{\boldsymbol{\phi}}_2\bigr)^2+2\,\mathbf{l}_1\,\mathbf{l}_2\,\dot{\boldsymbol{\phi}}_1\,\dot{\boldsymbol{\phi}}_2\cos\!\bigl(\boldsymbol{\phi}_1-\boldsymbol{\phi}_2\bigr)\Bigr]$

Our total kinetic energy would be

$\mathbf{T}=\frac{1}{2}(\mathbf{m}_1+\mathbf{m}_2)\bigl(\mathbf{l}_1\,\dot{\boldsymbol{\phi}}_1\bigr)^2+\frac{1}{2}\,\mathbf{m}_2\bigl(\mathbf{l}_2\,\dot{\boldsymbol{\phi}}_2\bigr)^2+\mathbf{m}_2\,\mathbf{l}_1\,\mathbf{l}_2\,\dot{\boldsymbol{\phi}}_1\,\dot{\boldsymbol{\phi}}_2\cos\!\bigl(\Delta\!\boldsymbol{\phi}\bigr)$

Onto our potential energy. There is only gravitational potential energy involved, so we can set the top as U = 0. U would be given by mgy. 

For Pendulum 1, $\mathbf{U}_1 = -\,\mathbf{m}_1\,g\,\mathbf{l}_1\cos\boldsymbol{\phi}_1$.

For Pendulum 2, $\mathbf{U}_2 = -\,\mathbf{m}_2\,g\,\bigl(\mathbf{l}_1\cos\boldsymbol{\phi}_1 + \mathbf{l}_2\cos\boldsymbol{\phi}_2\bigr)$

Then, our total potential energy is

$\mathbf{U}= -(\mathbf{m}_1+\mathbf{m}_2)\,g\,\mathbf{l}_1\cos\boldsymbol{\phi}_1-\mathbf{m}_2\,g\,\mathbf{l}_2\cos\boldsymbol{\phi}_2$

### Lagrangian

Now that we have our kinetic and potential energies, we can set up our Lagrangian. We can express the Lagrangian as the following, where $\Delta\!\boldsymbol{\phi}=\boldsymbol{\phi}_1-\boldsymbol{\phi}_2$:

$\mathbf{L}=\frac{1}{2}(\mathbf{m}_1+\mathbf{m}_2)\bigl(\mathbf{l}_1\,\dot{\boldsymbol{\phi}}_1\bigr)^2+\frac{1}{2}\,\mathbf{m}_2\bigl(\mathbf{l}_2\,\dot{\boldsymbol{\phi}}_2\bigr)^2+\mathbf{m}_2\,\mathbf{l}_1\,\mathbf{l}_2\,\dot{\boldsymbol{\phi}}_1\,\dot{\boldsymbol{\phi}}_2\cos\!\bigl(\Delta\!\boldsymbol{\phi}\bigr)+(\mathbf{m}_1+\mathbf{m}_2)\,g\,\mathbf{l}_1\cos\boldsymbol{\phi}_1+\mathbf{m}_2\,g\,\mathbf{l}_2\cos\boldsymbol{\phi}_2$

### Euler Lagrange Equations

Using Euler-Lagrange, we derive the following coupled equations of motion: 

$(\mathbf{m}_1+\mathbf{m}_2)\,\mathbf{l}_1\,\ddot{\boldsymbol{\phi}}_1
+\mathbf{m}_2\,\mathbf{l}_2\,\ddot{\boldsymbol{\phi}}_2\cos(\Delta\!\boldsymbol{\phi})
+\mathbf{m}_2\,\mathbf{l}_2\,\dot{\boldsymbol{\phi}}_2^{2}\sin(\Delta\!\boldsymbol{\phi})
+(\mathbf{m}_1+\mathbf{m}_2)\,g\,\sin\boldsymbol{\phi}_1=0$

$\mathbf{l}_1\,\ddot{\boldsymbol{\phi}}_1\cos(\Delta\!\boldsymbol{\phi})
+\mathbf{l}_2\,\ddot{\boldsymbol{\phi}}_2
-\mathbf{l}_1\,\dot{\boldsymbol{\phi}}_1^{2}\sin(\Delta\!\boldsymbol{\phi})
+g\,\sin\boldsymbol{\phi}_2=0$

Because these are coupled second order differential equations, we used algebra and substitution to solve for the angular accelerations.

Equation for $\ddot{\phi}_1$

$\ddot{\boldsymbol{\phi}}_1
=\frac{
-\,\mathbf{m}_2\,\mathbf{l}_2\,\dot{\boldsymbol{\phi}}_2^{2}\,\sin(\Delta\!\boldsymbol{\phi})
-\,\mathbf{m}_2\,\mathbf{l}_1\,\dot{\boldsymbol{\phi}}_1^{2}\,\sin(\Delta\!\boldsymbol{\phi})\cos(\Delta\!\boldsymbol{\phi})
-\,(\mathbf{m}_1+\mathbf{m}_2)\,g\,\sin\boldsymbol{\phi}_1
+\,\mathbf{m}_2\,g\,\sin\boldsymbol{\phi}_2\cos(\Delta\!\boldsymbol{\phi})
}{
\mathbf{l}_1\bigl(\mathbf{m}_1+\mathbf{m}_2\sin^2(\Delta\!\boldsymbol{\phi})\bigr)
}$

Equation for $\ddot{\phi}_2$

$\ddot{\boldsymbol{\phi}}_2
=\frac{
\mathbf{m}_2\,\mathbf{l}_2\,\dot{\boldsymbol{\phi}}_2^{2}\,\sin(\Delta\!\boldsymbol{\phi})\cos(\Delta\!\boldsymbol{\phi})
+(\mathbf{m}_1+\mathbf{m}_2)\,\mathbf{l}_1\,\dot{\boldsymbol{\phi}}_1^{2}\,\sin(\Delta\!\boldsymbol{\phi})
+(\mathbf{m}_1+\mathbf{m}_2)\,g\,\sin\boldsymbol{\phi}_1\cos(\Delta\!\boldsymbol{\phi})
-\,(\mathbf{m}_1+\mathbf{m}_2)\,g\,\sin\boldsymbol{\phi}_2
}{
\mathbf{l}_2\bigl(\mathbf{m}_1+\mathbf{m}_2\sin^2(\Delta\!\boldsymbol{\phi})\bigr)
}$

### Numerically Modeling a Double Pendulum System

Let's consider a double pendulum with $\mathit{l}_1$ = 5 m = $\mathit{l}_2$, $\mathit{m}_1$ = 2 kg = $\mathit{m}_2$, initial $\phi_1$ = $\frac{\pi}{2}$, initial $\phi_2$ = 0, and both pendulums starting from rest. We will use this system to plot the angular displacement over time, the state space, and the Poincare sections of both pendulums.

```python
import numpy as np
import matplotlib.pyplot as plt
from scipy.integrate import solve_ivp  # the differential equation integrator
from scipy.optimize import curve_fit  # for fitting the exponential decrease/
                                      #    increase between ICs
from scipy.interpolate import UnivariateSpline  # import the package to create
                                                #    a spline function
from matplotlib import animation #for animation 
from IPython.display import HTML  #for displaying animation inline
plt.rcParams["figure.figsize"] = (12,8)  # make figures big for easier viewing
                                         #    (width, height in inches)
# To revert to the small default size, uncomment the following line
# plt.rcParams["figure.figsize"] = plt.rcParamsDefault["figure.figsize"]
```

```python
### Initializing constants

# Apply limiting cases; when m2 = 0, should be a regular pendulum. When phi1 and phi2 are small angles, 
#should reduce to non-chaotic, linear motion through small angle approximation.

l1 = 1
l2 = 1
m1 = 1
m2 = 1
g = 9.81
```

```python
 #Initial conditions
phi_1_init = 1.5708
phi_2_init = 1.5708
ang_vel_1_init = 0
ang_vel_2_init = 0

def deriv(t_now, r_now, l1=5, l2=5, m1=2, m2=2):
    """
    Derivative function for the double pendulum system.
    r_now = [phi_1, ang_vel_1, phi_2, ang_vel_2]
    Returns: [ang_vel_1, alpha_1, ang_vel_2, alpha_2]
    """

    # Defining some values
    phi_1 = r_now[0]
    ang_vel_1 = r_now[1]
    phi_2 = r_now[2]
    ang_vel_2 = r_now[3]

    # Defining some trig values
    delta_phi = phi_1 - phi_2
    sin_phi_1 = np.sin(phi_1)
    cos_phi_1 = np.cos(phi_1)
    sin_phi_2 = np.sin(phi_2)
    cos_phi_2 = np.cos(phi_2)
    cos_delta = np.cos(delta_phi)
    sin_delta = np.sin(delta_phi)

    # Part of denominator in both equations
    denom = m1 + m2 * sin_delta**2

    # Making alpha_1
    num_alpha_1 = -m2*l2*ang_vel_2**2*sin_delta - m2*l1*ang_vel_1**2 *sin_delta*cos_delta - (m1 + m2)*g*sin_phi_1 + m2*g*sin_phi_2*cos_delta
    denom_alpha_1 = l1*denom
    alpha_1 = num_alpha_1/denom_alpha_1

    # Making alpha_2
    num_alpha_2 = m2*l2*ang_vel_2**2 *sin_delta*cos_delta + (m1 + m2)*l1*ang_vel_1**2 *sin_delta + (m1 + m2)*g*sin_phi_1*cos_delta - (m1 + m2) *g*sin_phi_2
    denom_alpha_2 = l2*denom
    alpha_2 = num_alpha_2/denom_alpha_2

    return np.array([ang_vel_1, alpha_1, ang_vel_2, alpha_2])
```

```python
### COMPUTE the numerical solution using solve_ivp
r_init = np.array([phi_1_init, ang_vel_1_init, phi_2_init, ang_vel_2_init])
t_span = (0, 50)
t_arr = np.linspace(0, 50, 1000)
r_soln = solve_ivp(deriv, t_span, r_init, args = (l1, l2, m1, m2, ), rtol = 1e-8, atol = 1e-8, t_eval = t_arr)
```

```python
### Plot phi vs t

fig, ax = plt.subplots()
ax.set_xlabel('t Time[s]')
ax.set_ylabel('phi')
ax.set_title('phi vs t')

phi_1 = r_soln.y[0]
phi_2 = r_soln.y[2]
time = r_soln.t
ax.plot(time, phi_1, color = 'blue')
ax.plot(time, phi_2, color = 'red')
ax.legend(['phi 1 vs t', 'phi 2 vs t'])
plt.show()

fig, ax2 = plt.subplots()
ax2.set_xlabel('x displacement')
ax2.set_ylabel('y displacement')
ax2.set_title('trajectory of double pendulum')

x1 = l1*np.sin(phi_1)
y1 = -l1*np.cos(phi_1)

x2 = x1 + l2*np.sin(phi_2)
y2 = y1 - l1*np.cos(phi_2)
ax2.plot(x1, y1, color = 'blue')
ax2.plot(x2, y2, color = 'red')
ax2.legend(['pendulum 1 trajectory', 'pendulum 2 trajectory'])
plt.show()
```

```python
time_per_frame = 0.4
num_frames = 500
dstep = np.max(t_arr)/(len(t_arr) - 1)
steps_per_frame = int(time_per_frame/dstep) 
print("Integrator time step (dt) = ", dstep)
print("Steps per frame (~{0:.2}/{1:.5}) = {2}".
      format(time_per_frame, dstep, steps_per_frame))


fig, ax = plt.subplots()
ax.set_aspect('equal')
lim = l1 + l2
ax.set_xlim(-lim, lim)
ax.set_ylim(-lim, lim)
ax.set_xlabel('x displacement')
ax.set_ylabel('y displacement')
ax.set_title('Double Pendulum with Animated Trajectories')

trajectory_1, = ax.plot([], [], '-', lw=1, alpha=0.6, color='blue')
trajectory_2, = ax.plot([], [], '-', lw=1, alpha=0.6, color='red')

m1dot,  = ax.plot([], [], 'ro', ms=6)
m2dot,  = ax.plot([], [], 'bo', ms=6)

def init():
    trajectory_1.set_data([], [])
    trajectory_2.set_data([], [])
    m1dot.set_data([], [])
    m2dot.set_data([], [])
    return trajectory_1, trajectory_2, m1dot, m2dot

def animate(i):
    j = min(int(i * steps_per_frame), len(t_arr) - 1)
    trajectory_1.set_data(x1[:j], y1[:j])
    trajectory_2.set_data(x2[:j], y2[:j])
    m1dot.set_data([x1[j]], [y1[j]])
    m2dot.set_data([x2[j]], [y2[j]])
    return trajectory_1, trajectory_2, m1dot, m2dot

anim = animation.FuncAnimation(
    fig, animate,
    init_func=init,
    frames=num_frames,
    blit=True,
    interval=100  
)

display(HTML(anim.to_jshtml()))
```

```python
### PLOT phi that is limited from -pi to pi versus t

fig, ax = plt.subplots()
ax.set_xlabel('t Time[s]')
ax.set_ylabel('phi 1')
ax.set_title('phi 1 bound between - pi and pi vs t')
new_phi_1 = (r_soln.y[0] + np.pi)%(2*np.pi) - np.pi
time = r_soln.t
ax.plot(time, new_phi_1, color = 'blue')
plt.show()

fig, ax2 = plt.subplots()
ax2.set_xlabel('t Time[s]')
ax2.set_ylabel('phi 2')
ax2.set_title('phi 2 bound between - pi and pi vs t')
new_phi_2 = (r_soln.y[2] + np.pi)%(2*np.pi) - np.pi
ax2.plot(time, new_phi_2, color = 'red')
plt.show()
```

```python
### Getting rid of horizontal lines (discontinuities) in state space

def no_horizontal_phi(phi, ang_vel):
    jumps = np.abs(np.diff(phi)) > np.pi
    indices = np.nonzero(jumps)[0] + 1 
    
    # Insert nan at jump locations
    changed_phi = np.insert(phi, indices, np.nan)
    changed_ang_vel = np.insert(ang_vel, indices, np.nan)
    
    return changed_phi, changed_ang_vel

```

```python
### State space orbits

fig, ax = plt.subplots()
ax.set_xlabel('phi 1')
ax.set_ylabel('angular velocity 1')
ax.set_title('phi 1 state space')
new_phi_1 = (r_soln.y[0] + np.pi)%(2*np.pi) - np.pi
ang_vel_1 = r_soln.y[1]
newer_phi_1, new_ang_vel_1 = no_horizontal_phi(new_phi_1, ang_vel_1)
ax.plot(newer_phi_1, new_ang_vel_1, color = 'blue')
plt.show()

fig, ax2 = plt.subplots()
ax2.set_xlabel('phi 2')
ax2.set_ylabel('angular velocity 2')
ax2.set_title('phi 2 state space')
new_phi_2 = (r_soln.y[2] + np.pi)%(2*np.pi) - np.pi
ang_vel_2 = r_soln.y[3]
newer_phi_2, new_ang_vel_2 = no_horizontal_phi(new_phi_2, ang_vel_2)
ax2.plot(newer_phi_2, new_ang_vel_2, color = 'red')
plt.show()
```

# Poincare Sections

For the double pendulum system we have been working with, we are going to plot the corresponding Poincare sections for both the top and bottom pendulums. For our purposes, we will work with the same time span (0 to 50 seconds) and time step (1000 points, or 0.05 s). We plot the Poincare sections using spline fits, which we imported earlier.

```python
### Poincare Sections

# Initializing phi 1~2, angular velocity 1~2
new_phi1 = (r_soln.y[0] + np.pi)%(2*np.pi) - np.pi
new_phi2 = (r_soln.y[2] + np.pi)%(2*np.pi) - np.pi
t = r_soln.t
phi_1_mod = (r_soln.y[0] + np.pi)%(2*np.pi) - np.pi
ang_vel_1_mod = r_soln.y[1]
phi_1, ang_vel_1 = no_horizontal_phi(phi_1_mod, ang_vel_1_mod)

phi_2_mod = (r_soln.y[2] + np.pi)%(2*np.pi) - np.pi
ang_vel_2_mod = r_soln.y[3]
phi_2, ang_vel_2 = no_horizontal_phi(phi_2_mod, ang_vel_2_mod)

### USE spline to find the value of phi and angular velocity
t_arr = np.linspace(0, 50, 1000)

spl_1a = UnivariateSpline(t, new_phi_1, k = 5, s = 0)
phi_1_fit = spl_1a(t_arr)

spl_1b = UnivariateSpline(t, ang_vel_1, k = 5, s = 0)
ang_vel_1_fit = spl_1b(t_arr)

spl_2a = UnivariateSpline(t, new_phi_2, k = 5, s = 0)
phi_2_fit = spl_2a(t_arr)

spl_2b = UnivariateSpline(t, ang_vel_1, k = 5, s = 0)
ang_vel_2_fit = spl_2b(t_arr)

### ADD code to plot the Poincar√© sections of both pendulums
fig, ax = plt.subplots()
ax.set_xlabel('phi 1 [rad]')
ax.set_ylabel('angular velocity 1 [rad/s]')
ax.set_title('Poincare section for Top Pendulum')
ax.axis([-np.pi, np.pi, -5, 5])
ax.plot(phi_1_fit, ang_vel_1_fit, '.', color = 'b')
plt.show()

fig, ax2 = plt.subplots()
ax2.set_xlabel('phi 2 [rad]')
ax2.set_ylabel('angular velocity 2 [rad/s]')
ax2.set_title('Poincare section for Bottom Pendulum')
ax2.axis([-np.pi, np.pi, -5, 5])
ax2.plot(phi_2_fit, ang_vel_2_fit, '.', color = 'r')
ax2.set_xlim(-4, 4)
plt.show()
```

### The Next Steps: Limiting Cases

To prove our derivation and thus our data is correct, we need to test it on some limiting cases to show that it behaves as expected. For example, we know that under very small angles, like ${\pi}$/6, for both of our ${\phi}$, the behavior our pendulum should undergo should be non-chaotic. Additionally, if $\mathbf{m}_2$ is 0 kg, our double pendulum system must act like a simple pendulum. In these next few cells we will test these.

```python
# step 1: Small angles
phi_1_init_2 = np.pi/6
phi_2_init_2 = np.pi/6
ang_vel_1_init_2 = 0
ang_vel_2_init_2 = 0

r_init_2 = np.array([phi_1_init_2, ang_vel_1_init_2, phi_2_init_2, ang_vel_2_init_2])
t_span = (0, 50)
t_arr = np.linspace(0, 50, 1000)
r_soln_2 = solve_ivp(deriv, t_span, r_init_2, args = (l1, l2, m1, m2, ), rtol = 1e-8, atol = 1e-8, t_eval = t_arr)
```

```python
#step 2: Graphing 
fig, ax = plt.subplots()
ax.set_xlabel('t Time[s]')
ax.set_ylabel('phi')
ax.set_title('phi vs t')

phi_1_2 = r_soln_2.y[0]
phi_2_2 = r_soln_2.y[2]
time = r_soln_2.t
ax.plot(time, phi_1_2, color = 'blue')
ax.plot(time, phi_2_2, color = 'red')
ax.legend(['phi 1 vs t', 'phi 2 vs t'])
plt.show()

fig, ax2 = plt.subplots()
ax2.set_xlabel('x displacement')
ax2.set_ylabel('y displacement')
ax2.set_title('trajectory of double pendulum')

x1 = l1*np.sin(phi_1_2)
y1 = -l1*np.cos(phi_1_2)

x2 = x1 + l2*np.sin(phi_2_2)
y2 = y1 - l1*np.cos(phi_2_2)
ax2.plot(x1, y1, color = 'blue')
ax2.plot(x2, y2, color = 'red')
ax2.legend(['pendulum 1 trajectory', 'pendulum 2 trajectory'])
plt.show()
```

As we can see, both trajectories oscillate back and forth non-chaotically just as expected. Now onto the other test.

```python
# step 3: no bottom pendulum
phi_1_init_3 = 1.5708
phi_2_init_3 = 1.5708
ang_vel_1_init_3 = 0
ang_vel_2_init_3 = 0

l2_2 = 1
m2_2 = 0

r_init_3 = np.array([phi_1_init_3, ang_vel_1_init_3, phi_2_init_3, ang_vel_2_init_3])
t_span = (0, 50)
t_arr = np.linspace(0, 50, 1000)
r_soln_3 = solve_ivp(deriv, t_span, r_init_3, args = (l1, l2_2, m1, m2_2, ), rtol = 1e-8, atol = 1e-8, t_eval = t_arr)
```

```python
fig, ax = plt.subplots()
ax.set_xlabel('t Time[s]')
ax.set_ylabel('phi')
ax.set_title('phi vs t')

phi_1_3 = r_soln_3.y[0]
phi_2_3 = r_soln_3.y[2]
time = r_soln_3.t
ax.plot(time, phi_1_3, color = 'blue')
ax.plot(time, phi_2_3, color = 'red')
ax.legend(['phi 1 vs t', 'phi 2 vs t'])
plt.show()

fig, ax2 = plt.subplots()
ax2.set_xlabel('x displacement')
ax2.set_ylabel('y displacement')
ax2.set_title('trajectory of double pendulum')

x1 = l1*np.sin(phi_1_3)
y1 = -l1*np.cos(phi_1_3)

x2 = x1 + l2*np.sin(phi_2_3)
y2 = y1 - l1*np.cos(phi_2_3)
ax2.plot(x1, y1, color = 'blue')
ax2.plot(x2, y2, color = 'red')
ax2.legend(['pendulum 1 trajectory', 'pendulum 2 trajectory'])
plt.show()
```

One would think this trajectory is chaotic by looking at pendulum 2. However, as m2 equals 0, the blue trajectory represents a massless string. Focusing on only the red trajectory, we can see that the double pendulum system with only mass m1 behaves like a simple pendulum, which is as expected.
